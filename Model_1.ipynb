```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("data/cleaned_amazon_reviews_final.csv")
print(df.head())
```

```python
# View column names
print(df.columns)

# Rename columns if needed to standardize
df = df[['text', 'rating']].dropna()
df = df[df['rating'].isin(['positive', 'neutral', 'negative'])]
```

```python
label_encoder = LabelEncoder()
df['label'] = label_encoder.fit_transform(df['rating'])
```

```python
# Check if dataframe is empty before splitting
if df.empty:
    print("DataFrame is empty! Please check your data loading step and file path.")
else:
    train_texts, val_texts, train_labels, val_labels = train_test_split(
        df['text'].tolist(), df['label'].tolist(), test_size=0.2, random_state=42
    )
```
